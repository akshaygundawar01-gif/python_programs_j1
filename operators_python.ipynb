3.comparision operators

5.membership operators

```python
# arthematic operators -+,-,*,/,//,%
a=10
b=20
c=20
d=30
print(a+b)
print(b-c)
print(b*d)
print(d/a)
print(d//a)
print(d%a)
```

```python
#assigment operators -+=,-=,*=,/=,//=,%=
a=100
a+=50
print(a)
```

```python
b=200
b-=400
print(b)
```

```python
c=50
c*=100
print(c)
```

```python
# comparsion operators ->,<,>=,<=,==,!=
a=10
b=20
c=20
d=30
print(a>b)
print(b==c)
print(b<c)
print(b!=c)
print(b>=a)
print(b<=a)
```

```python
# membership operators -in,not, in 
# validating member are present in that sequence or group
names=["vikas','shivagi","sidh","chndra"]
print("vikas" in names)
print("shivaji" not in names)
print("vikas" not in names)
print("vikram" in names)
```

```python
# identity operators -is, is not
a=10000
b=10000
print(a is b)
print(id(a))
print(id(b))
```
